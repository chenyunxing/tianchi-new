```python
#coding:utf-8
import pandas as pd
import time

from sklearn.metrics import log_loss
def time2cov(time_):
    '''
    时间是根据天数推移，所以日期为脱敏，但是时间本身不脱敏
    :param time_: 
    :return: 
    '''
    return time.strftime("%Y-%m-%d %H:%M:%S",time.localtime(time_))
def pre_process(data):
    '''
    :param data: 
    :return: 
    '''

    print('预处理')
    print('context_timestamp_ing')
    data['context_timestamp1'] = data['context_timestamp'].apply(time2cov)
    print('time')
    data['context_timestamp1'] = pd.to_datetime(data['context_timestamp1'])
    data['day'] = data['context_timestamp1'].dt.day
    del data['context_timestamp1']
    return data
```

```python
print('train')
train = pd.read_table('train.txt',sep=' ',index_col=False)
```

```python
train = pre_process(train)
```
