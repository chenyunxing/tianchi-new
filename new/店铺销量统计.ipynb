```python
print('train')
train = pd.read_table('train.txt',sep=' ',index_col=False)
```

```python
test_a = pd.read_table('test.txt',sep=' ',index_col=False) 
```

```python
train[['is_trade','shop_id']].groupby('shop_id').sum()
```

```python
temp = pd.DataFrame()
```

```python
temp['s_pv'] = (train[['shop_id','is_trade']].groupby('shop_id').sum()).reset_index()[['is_trade']]
```

```python
temp['shop_id'] = (train[['shop_id','is_trade']].groupby('shop_id').sum()).reset_index()[['shop_id']]
```

```python
temp
```

```python
train = pd.merge(train[['instance_id','shop_id']],temp[['shop_id','s_pv']],on=['shop_id'],how='left')
test_a = pd.merge(test_a[['instance_id','shop_id']],temp[['shop_id','s_pv']],on=['shop_id'],how='left')
```
