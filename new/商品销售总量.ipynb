```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
from scipy.stats import norm, skew #加载一些统计的库
import time
import gc
import math
```

```python
print('train')
train = pd.read_table('train.txt',sep=' ',index_col=False)
```

```python
test_a = pd.read_table('test.txt',sep=' ',index_col=False) 
```

```python
train[['is_trade','item_id']].groupby('item_id').sum()
```

```python
temp = pd.DataFrame()
```

```python
temp['i_pv'] = (train[['item_id','is_trade']].groupby('item_id').sum()).reset_index()[['is_trade']]
```

```python
temp['item_id'] = (train[['item_id','is_trade']].groupby('item_id').sum()).reset_index()[['item_id']]
```

```python
temp
```

```python
train = pd.merge(train[['instance_id','item_id']],temp[['item_id','i_pv']],on=['item_id'],how='left')
test_a = pd.merge(test_a[['instance_id','item_id']],temp[['item_id','i_pv']],on=['item_id'],how='left')
```

```python
test_a = test_a.fillna(0)
```
