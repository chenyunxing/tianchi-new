```python
def time2cov(time_):
    '''
    时间是根据天数推移，所以日期为脱敏，但是时间本身不脱敏
    :param time_: 
    :return: 
    '''
    return time.strftime("%Y-%m-%d %H:%M:%S",time.localtime(time_))
```

```python
def pre_process(data):
    '''
    :param data: 
    :return: 
    '''

    print('item_category_list_ing')
    for i in range(3):
        data['category_%d'%(i)] = data['item_category_list'].apply(
            lambda x:x.split(";")[i] if len(x.split(";")) > i else " "
        )
    del data['item_category_list']

    print('item_property_list_ing')
    for i in range(3):
        data['property_%d'%(i)] = data['item_property_list'].apply(
            lambda x:x.split(";")[i] if len(x.split(";")) > i else " "
        )
    del data['item_property_list']

    return data
```

```python
data = pd.read_table('train.txt',sep=' ',index_col=False)
data2 = pd.read_table('test.txt',sep=' ',index_col=False) 
data3 = pd.read_table('test1.txt',sep=' ',index_col=False) 
```

```python
data = pre_process(data)
data2 = pre_process(data2)
data3 = pre_process(data3)
```

```python
all_data = data.append(data2)
all_data = all_data.append(data3)
```

```python
user_gender_click = all_data[['user_gender_id','category_1','instance_id']].groupby(['user_gender_id','category_1']).count()/data[['user_gender_id','instance_id']].groupby(['user_gender_id']).count()
```

```python
user_gender_click = user_gender_click.rename(columns=({'instance_id':'user_gender_click'}))
```

```python
user_gender_click = user_gender_click.reset_index()
```

```python
user_gender_click_data = pd.merge(data,user_gender_click,on=['user_gender_id','category_1'],how='left')
```

```python
user_gender_click_data2 = pd.merge(data2,user_gender_click,on=['user_gender_id','category_1'],how='left')
```

```python
user_gender_click_data3 = pd.merge(data3,user_gender_click,on=['user_gender_id','category_1'],how='left')
```
